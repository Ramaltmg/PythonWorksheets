```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

A_inv = np.linalg.inv(A)
I = np.dot(A, A_inv)

print("A * A⁻¹ =\n", I)
```

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

AB = np.dot(A, B)
BA = np.dot(B, A)

print("A*B =\n", AB)
print("B*A =\n", BA)
print("A*B == B*A ?", np.array_equal(AB, BA))
```

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

AB_T = (np.dot(A, B)).T
B_T_A_T = np.dot(B.T, A.T)

print("(AB)^T =\n", AB_T)
print("B^T * A^T =\n", B_T_A_T)
print("(AB)^T == B^T*A^T ?", np.array_equal(AB_T, B_T_A_T))
```

```python
import numpy as np

A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

B = np.array([-1, -3, 9])

X = np.linalg.solve(A, B)

print("Solution [x, y, z]:", X)
```

```python
import numpy as np

A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

B = np.array([-1, -3, 9])

A_inv = np.linalg.inv(A)

X = np.dot(A_inv, B)

print("Solution [x, y, z]:", X)
```
